# **Step 4 Merge the master data frame and yearly data based on Key column**

```python
mergeddf2020= pd.merge(
    dfmaster_desc,
    df_2020,
    how="inner",
    on="HS6-SH6",
    left_on=None,
    right_on=None,
    left_index=False,
    right_index=False,
    sort=True,
    copy=True,
    indicator=False,
    validate=None,
)
```
